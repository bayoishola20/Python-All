```python
profile = io.imread('./Data/profile.jpg')
type(profile)
```

```python
plt.imshow(profile) # display picture
```

```python
profile.shape
```

```python
plt.imshow(profile[::-1]) # [::-1] will display image upside-down (reverse the rows in the image)
```

```python
plt.imshow(profile[:, ::-1]) # mirror image - reverse column from last
```

```python
plt.imshow(profile[::-1, :])
```

```python
plt.imshow(profile[:150, 100:]) # get only my head
```

```python
plt.imshow(profile[::2, ::2]) # hald the size of the image
```

```python
photo_sin = np.sin(profile)
photo_sin
```

```python
# data (260, 260, 3) => Axis 0: 260, Axis 1: 260, Axis 2: 3
print("Sum: ", np.sum(profile))
print("Prod: ", np.prod(profile))
print("Mean: ", np.mean(profile))
print("Median: ", np.median(profile))
print("STD: ", np.std(profile))
print("Variance: ", np.var(profile))
print("Minimum: ", np.min(profile))
print("Maximum: ", np.max(profile))
print("ArgMin: ", np.argmin(profile)) # Returns the indices of the minimum values along an axis.
print("ArgMax: ", np.argmax(profile)) # Returns the indices of the minimum values along an axis.
```

```python
profile_masked = np.where(profile > 150, 150, 0) # where profile pic is > 150, replaces with 255, otherwise 0
plt.imshow(profile_masked)
```

```python
plt.imshow(profile[:, :, 0].T) # T for transform
```
